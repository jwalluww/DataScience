Variable Clustering - Variable Reduction Technique to avoid multicollinearity

Import two packages, including Var Clus Hi

```python
import pandas as pd
from varclushi import VarClusHi
```

Read in dataset of your choice. Using the Fifa dataset here...

```python
data=pd.read_csv('C:\\Users\\wallj\\DS_Projects\\fifa\\fifa18-all-player-statistics\\2019\\data.csv')
```

Keep only variables you'd like to use

```python
df = data[[
'Crossing'
,'Finishing'
,'HeadingAccuracy'
,'ShortPassing'
,'Volleys'
,'Dribbling'
,'Curve'
,'FKAccuracy'
,'LongPassing'
,'BallControl'
,'Acceleration'
,'SprintSpeed'
,'Agility'
,'Reactions'
,'Balance'
,'ShotPower'
,'Jumping'
,'Stamina'
,'Strength'
,'LongShots'
,'Aggression'
,'Interceptions'
,'Positioning'
,'Vision'
,'Penalties'
,'Composure'
,'Marking'
,'StandingTackle'
,'SlidingTackle'
,'GKDiving'
,'GKHandling'
,'GKKicking'
,'GKPositioning'
,'GKReflexes'
]]
```

No nulls allowed, fill or drop now

```python
df.dropna(inplace=True)
```

Create the varclus object with inputs dataset, eigen value and max clusters

```python
# Change Eigen Value to alter clusters - higher = larger clusters = fewer varaibles at end
varclusdf = VarClusHi(df,maxeigval2=0.75,maxclus=None)
```

Run the actual varclus function

```python
%%time
varclusdf.varclus()
```

This will give the cluster number, number of variables per cluster, and other information by cluster

```python
varclusdf.info
```

This gives each variable, it's r-square for it's cluster and the next closest, as well as the ratio between those

```python
varclusdf.rsquare
```

Use the variable-specific output abhove to build a new df

```python
df_varclus = pd.DataFrame(data=varclusdf.rsquare)
```

Build a df using just the cluster-level information for safe keeping

```python
df_varclus_info = pd.DataFrame(data=varclusdf.info)
```

Sort the variable-specific df by clustser and r-square ratio and choose the lowest of the r-sqare ratios by cluster

```python
df_sorted=df_varclus.groupby(['Cluster'],sort=False)['RS_Ratio'].transform(min)==df_varclus['RS_Ratio']
```

List of new variables

```python
vars_to_keep=df_varclus[df_sorted]['Variable'].tolist()
```

New dataframe with only keeper variables

```python
df1=df[vars_to_keep]
```
